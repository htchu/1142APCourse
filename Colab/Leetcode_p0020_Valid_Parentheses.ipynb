# Leetcode-p0020-Valid Parentheses

## Method 1: 2 for-loops

```python
nums = [2,7,11,15]
target = 9
n=len(nums)
for i in range(n-1):
  for j in range(i+1, n):
      if nums[i]+nums[j]==target:
          print([i,j])
```

Method 2: dictionary

```python
nums = [2,7,11,15]
target = 9
n=len(nums)
reversed_complement=dict()
for i in range(n):
  rc=target-nums[i]
  reversed_complement[i]=rc
for i in range(n):
  rc=target-nums[i]
  if rc in reversed_complement:
    print([reversed_complement[rc],i])
```
