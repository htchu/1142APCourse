# Leetcode-p0020-Valid Parentheses

## Method 1: List implementation of stack

```python
s = "()[]{}"
stack = []
pairs = {')': '(', ']': '[', '}': '{'}
for char in s:
    if char in pairs.values():
      stack.append(char)
    elif char in pairs:
       if not stack or stack[-1] != pairs[char]:
          print(False)
       stack.pop()
if not stack:
  print(True)
else:
  print(False)

```
